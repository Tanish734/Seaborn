# Figure Level Function -> Catplot

```python
# strip plot
# axis level
sns.stripplot(data = tips, x = 'day', y = 'total_bill' )
```

```python
sns.stripplot(data = tips, x = 'day', y = 'total_bill',jitter=False)
```

```python
# Using catplot 
# figure level plot functions

sns.catplot(data = tips , x = 'day',y = 'total_bill', kind = 'strip')
```

```python
sns.catplot(data = tips , x = 'day',y = 'total_bill', kind = 'strip',jitter = False)
```

```python
sns.catplot(data = tips , x = 'day',y = 'total_bill', kind = 'strip',jitter = 0.2, hue = 'sex')
```

```python
#swarmplot
sns.catplot(data = tips , x= 'day',y = 'total_bill', kind = 'swarm', color = 'red')
```

```python
sns.swarmplot(data = tips , x= 'day',y = 'total_bill'  , color = 'red', hue = 'sex')
```

# Boxplot
    ## . A boxplot is a standardized way of displaying the distribution of data based on a five number summary
    ##  ("minimum", first quatile[Q1], median,third quatile[Q3] and maximum)
    ##  it can tell ypu about your outlier and what their values are . Boxplot can also tell if you data is symmetrical ,
    ##  how tightly your data is grouped and  if  and how data is skewed.
    

```python
# Box plot
sns.boxplot(data = tips, x = 'day',y = 'total_bill', hue = 'sex')
```

```python
sns.catplot(data = tips, x = 'day',y = 'total_bill', hue = 'sex', kind = 'box')
```

```python
sns.boxplot(data = tips , y = 'total_bill')
```

# violinplot = (Boxplot + Kdeplot)

```python
sns.violinplot(data = tips ,x = 'day',y = 'total_bill')
```

```python
sns.catplot(data = tips ,x = 'day',y = 'total_bill', kind = 'violin', hue = 'sex',split = True)
```

```python
# barplot
import numpy as np
sns.barplot(data = tips , x = 'sex', y = 'total_bill',estimator = np.std, ci = None)
```

# When there are # multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

```python
sns.catplot(data = tips , x = 'sex', y = 'total_bill',estimator = np.std, kind = 'bar')
```

```python
# point plot
sns.catplot(data = tips , x = 'sex', y = 'total_bill',estimator = np.mean, kind = 'point', hue = 'smoker',ci = None)
```

```python
sns.pointplot(data = tips , x = 'sex', y = 'total_bill' , hue = 'smoker', estimator = np.mean)
```

```python
sns.pointplot(data = tips , x = 'sex', y = 'total_bill', hue = 'smoker',ci = None)
```

```python
# coount
sns.countplot(data = tips, x = 'sex', hue = 'smoker')
```

# A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable

```python
# faceting using catplot

sns.catplot(data = tips, x = 'sex', y = 'total_bill', col = 'smoker', kind = 'box', row = 'time')
```

# Regression Plots
# regplot
# lmplot
# In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
# axes level figure
# hue parameter not available
sns.regplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
sns.lmplot(data = tips, x = 'total_bill', y = 'tip',hue = 'sex')
```

```python
# residual plot

sns.residplot(data = tips , x = 'total_bill', y = 'tip')
```

# Asecond way to plot facet plot -> FacetGrid

```python
# figure level -> relplot, dislplot , catplot , lmplot

sns.catplot(data = tips,x = 'sex', y = 'total_bill', kind = 'violin', col = 'day', row = 'time')
```

```python
g =sns.FacetGrid(data = tips, col = 'day', row = 'time')
g.map(sns.violinplot, 'sex', 'total_bill')
```

```python
g =sns.FacetGrid(data = tips, col = 'day', row = 'time')
g.map(sns.boxplot, 'sex', 'total_bill')
```

# pairplot Pairwise Relationship(PairGrid vs Pairplot)

```python
iris
```

```python
# pariplot
sns.pairplot(iris,hue = 'species')
```

```python
# pairgrid
g = sns.PairGrid(data =iris, hue = 'species')
# g.map
g.map(sns.scatterplot )
```

```python
g =sns.PairGrid(data = iris)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g =sns.PairGrid(data = iris)
g.map_diag(sns.violinplot)
g.map_offdiag(sns.scatterplot)
```

```python
g =sns.PairGrid(data = iris)
g.map_diag(sns.boxplot)
g.map_offdiag(sns.histplot)
```

```python
g =sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
g =sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
g =sns.PairGrid(data = iris, hue = 'species',vars = ['sepal_length','sepal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
