```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
# scatter
# axis level functions
sns.scatterplot(data = tips,
                x = 'total_bill',
                y = 'tip', 
                hue = 'sex',
                style = 'time',
                size = 'size'
               )
```

```python
# relplot -> figure level -> square shape
sns.relplot(data = tips,
            x = 'total_bill',
            y = 'tip',
            kind = 'scatter',
            hue = 'sex',
            style = 'time',
            size = 'size'
           )
```

# Distribution plot 
    ## . used for univariate analysis
    ## . used to find out the distribution
    ## . Range of the observation
    ## . Central Tendency
    ## . is the data bimodel
    ## . Are there outliers
# Plot under distribution plot 
    ## . hisplot
    ## . kdeplot
    ## . rugplot

# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot

```python
# plotting univariate histogram
sns.histplot(data = tips, x = 'total_bill')
```

```python
sns.displot(data = tips , x = 'total_bill', kind = 'hist')
```

```python
# bins parameter
sns.displot(data = tips , x = 'total_bill', kind = 'hist', bins = 20)
```

```python
# It's aslo possible to visulize the distribution of a variable using the logic of a histogram.
# Discrete bins are automatically  set for categorical variable
sns.displot(data = tips , x = 'day', kind = 'hist')
```

```python
sns.displot(data = tips , x = 'tip', kind = 'hist',hue = 'sex')
```

```python
sns.displot(data = tips , x = 'tip', kind = 'hist',hue = 'sex', element = 'step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data = titanic , x = 'age', kind = 'hist',hue = 'survived', element = 'step')
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data = titanic , x = 'age', kind = 'hist',col = 'sex', element = 'step')
```

# kdeplot 
# Rather than using discrete bins , a kde plot smooth the observation with a gaussian kernel , producing a continuous density estimate

```python
sns.kdeplot(data = tips , x = 'total_bill' )
```

```python
# hue fill
sns.displot(data = tips , x = 'total_bill', kind = 'kde' )
```

```python
sns.displot(data = tips , x = 'total_bill', kind = 'kde', hue = 'sex' , fill = True)
```

# Rugplot
# Plot marginal distributoin by drawing ticks along the x and y axes.
# This function is intended to complement other plots by showing the location of indiviual observations in an unobstusive way

sns.kdeplot(data = tips , x = 'total_bill')
sns.rugplot(data = tips , x = 'total_bill')

# Matrix Plot
    ## . Heatmap
    ## . Clustermap

```python
# heatmap


# Plot rectangular data as a color- encoded matrix

temp_df = gap.pivot(index = 'country', columns = 'year', values = 'lifeExp')
```

```python
# axes level function
plt.figure(figsize = (20,20))
sns.heatmap(temp_df)
```

```python
temp_df= gap[gap['continent'] == 'Europe'].pivot(index = 'country', columns = 'year', values = 'lifeExp')
plt.figure(figsize = (20,20))
sns.heatmap(temp_df, annot=True, linewidth = 0.5, cmap = 'crest')
```

```python
# clustermap
# plot a matrix dataset as a hierarchically - clusterd heatmap
# This function require sciply to be available


iris= px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
